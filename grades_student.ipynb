This program takes input from, presumably, an instructor, that includes each student's name and grades (4 test scores)
That information gets slurped into a dictionary where each element is transformed into a list. Next, we explore taking the average for each student (all 4 test scores), class average (based on each student's average) and then cumulative averages for each test; i.e. average for test1, test2, etc. This information is plotted via plotly to see the variance relative to the mean. In the end, a dataframe is created that pulls in the aforementioned information, as well as creating new columns based on an applied curve and adding the letter grade based off the final adjusted grade.

```python
#create empty dictionary
scores = {}
```

```python
for i in range(12): #number of students
    stu_name = input("Please enter the student's name: \n")
    for i in range(1): #number of iterations to grabe the test score
        grade1 = int(input(f"Please enter {stu_name}'s 1st test score: \n"))
        grade2 = int(input(f"Please enter {stu_name}'s 2nd test score: \n"))
        grade3 = int(input(f"Please enter {stu_name}'s 3rd test score: \n"))
        grade4 = int(input(f"Please enter {stu_name}'s 4th test score: \n"))
    scores[stu_name.title()] = grade1, grade2, grade3, grade4
```

```python
student_avg = [] #student's name with their average
raw_avg = [] #just the averages all combined into a list
for i in grades:
    raw_avg.append(round(np.mean(i[1]),1))
    student_avg.append(i[0]+': '+ str(round(np.mean(i[1]),1)))
```

```python
x = [i[1] for i in grades]
m = []
for i in x:
    m.append(list(i))
```

```python
grade_df['Student']= names
grade_df['Scores']= [i for i in m]
grade_df['Average']=student_averages
```

```python
grade_df
```

```python
#reminder of the class average so we can apply a curve
class_avg
```

```python
lowest_avg = min(grade_df['Average'])
highest_avg = max(grade_df['Average'])
```

```python
print(f'The lowest average in the class is {lowest_avg}, and the highest is {highest_avg}.\nBased off of this information, we can apply a curve to ensure that the lowest score meets the class average')
```

```python
diff= round(class_avg - lowest_avg,1)
```

```python
grade_df['Adjusted Grade'] = grade_df['Average'] + diff
```

```python
grade_df
```

```python
#add letter grade
grades = []
for i in grade_df['Adjusted Grade']:
    grades.append(i)
    
letters = []
for i in grades:
    if i > 97:
        letters.append('A+')
    elif i < 97 and i > 92:
        letters.append('A')
    elif i > 89.9 and i < 93:
        letters.append('A-')
    elif i > 86.9 and i < 90:
        letters.append('B+')
    elif i > 82 and i < 87:
        letters.append('B')
    elif i > 79 and i < 83:
        letters.append('B-')
    elif i > 76 and i < 80:
        letters.append('C+')
    elif i > 72 and i < 77:
            letters.append('C')
    else:
        if i > 69 and i < 73:
            letters.append('C-')
```

```python
letters
```

```python
grade_df['Final Letter Grade']= letters
```
